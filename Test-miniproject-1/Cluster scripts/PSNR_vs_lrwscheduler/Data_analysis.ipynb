```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice

from torch import nn
from torch.nn import functional as F

import importlib
import sys 
import getpass
import glob
```

```python
files = np.array(glob.glob("Data_new/*"))

lrs = np.array([ float(f.split("_")[2].split("=")[-1]) for f in files])
bs = 16

files = files[np.argsort(lrs)]
lrs = np.sort(lrs)
```

```python
for lr, f in zip(lrs, files):
    nim  = np.loadtxt(f)[:,0]
    psnr = np.loadtxt(f)[:,-1]
    plt.plot(nim/(bs*150), psnr, '-', label="%.0e"%(lr))

plt.legend(ncol=2, title=r"Learning rates")

plt.axhline(23, ls='--', color='grey', alpha=0.6)
plt.axhline(24, ls='--', color='grey', alpha=0.6)
plt.axhline(25, ls='--', color='grey', alpha=0.6)
plt.ylim(18,)

plt.xscale("log")
plt.xlabel(r"# images / $B$")
plt.ylabel("PSNR");
#plt.xlim(-100000,0.2e7)
```
