```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice

from torch import nn
from torch.nn import functional as F

import importlib
import sys 
import getpass
import glob

sys.path.append('/home/luca_ltpn/DL-Fleuret/Proj_308427_348143_XXXXXX')
sys.path.append('/home/luca_ltpn/DL-Fleuret')
from Miniproject_1.model import *
```

```python
files = np.array(glob.glob("Data/*"))

pdo = np.array([ float(f.split("_")[1].split("=")[-1]) for f in files])
bs  = 16

files = files[np.argsort(pdo)]
pdo = np.sort(pdo)
```

```python
for p, f in zip(pdo, files):
    nim  = np.loadtxt(f)[:,0]
    psnr = np.loadtxt(f)[:,-1]
    plt.plot(nim/(bs*150), psnr, '-', label="%.1e"%(p))

plt.legend(ncol=2, title=r"Dropout probability")

plt.axhline(23, ls='--', color='grey', alpha=0.6)


plt.xscale("log")
plt.xlabel(r"# images / $B$")
plt.ylabel("PSNR");
#plt.xlim(-100000,0.2e7)
```
